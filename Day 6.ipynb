```python
import typing

# Create world map
world = [list(line) for line in values.split()]


### Part 1

# Function to progress through the world
def sleigh_on(xAugment: int, yAugment: int) -> int:
    lineLength = len(world[0])
    totalHeight = len(world)
    x = 0
    y = 0
    trees = 0
    while True:
        #print('x: {} y: {}'.format(x,y))
        if y > lineLength - 1:
            y = y % (lineLength)
            #print('RESET: x :{} y :{}'.format(x,y))
        if x > totalHeight - 1:
            #print('Done')
            break
        try:
            if world[x][y] == '#':
                #print('Found a tree!')
                trees += 1
            x += xAugment
            y += yAugment
        except:
            print('exceeded world bounds...')
            break
    return trees
            
print('Total trees: {}'.format(sleigh_on(1,3)))
```

```python
### Part 2
import numpy as np
product = np.prod([sleigh_on(*z) for z in [(1,1),(1,3),(1,5),(1,7),(2,1)]])
print('Run product: {}'.format(product))
```
