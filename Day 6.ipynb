```python
### Part 2 
import typing
from typing import List, Set
# Get the set of common values for all people

def get_unique_letters_all(answer: List) -> Set:
    letters = set()
    # Construct the letters used by any person
    for item in answer:
        if isinstance(item, list):
            tItem = set(item)
            letters = letters | tItem
        elif isinstance(item, str):
            letters = letters | set(item)
    # Now check if each letter exists in all items
    finalLetters = []
    for letter in letters:
        if all([letter in item for item in answer]):
            finalLetters.append(letter)
    return finalLetters
    
answer = sum([len(get_unique_letters_all(item)) for item in map(list,[group.split() for group in values.split('\n\n')])])
print(f'Sum of Custom Counts: {answer}')
```
