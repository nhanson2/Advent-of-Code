```python
### Part 1
import typing
from typing import List
# Clean input
instructionsFinal = [line.split() for line in values.split('\n')]
```

```python
def execute_set(instructionsFinal: List) -> int:
    # Initialize accumulator
    accumulator = 0
    # Iterate over the instructions
    instructIndex = 0
    runInstructions = set()
    while True:
        currInstruct = instructionsFinal[instructIndex]
        # Check if we have already run this command
        if instructIndex in runInstructions:
            break
        else:
            runInstructions.add(instructIndex)
        # Perform the command
        if currInstruct[0] == 'acc':
            if currInstruct[1][0] == '-':
                accumulator -= int(currInstruct[1][1:])
            if currInstruct[1][0] == '+':
                accumulator += int(currInstruct[1][1:])
            instructIndex += 1
        elif currInstruct[0] == 'jmp':
            if currInstruct[1][0] == '-':
                instructIndex -= int(currInstruct[1][1:])
            if currInstruct[1][0] == '+':
                instructIndex += int(currInstruct[1][1:])
        elif currInstruct[0] == 'nop':
            instructIndex += 1
    return accumulator

print(f'Value of accumulator at repeat {execute_set(instructionsFinal)}')
```

```python
### Part 2
from copy import deepcopy
def execute_set_valid(instructionsFinal: List) -> int:
    # Initialize accumulator
    accumulator = 0
    # Iterate over the instructions
    instructIndex = 0
    runInstructions = set()
    while True:
        # If the next instruction follows the final sequential instruction, we have success
        if instructIndex > len(instructionsFinal) - 1:
            return accumulator
        currInstruct = instructionsFinal[instructIndex]
        # Check if we have already run this command
        if instructIndex in runInstructions:
            return 0
        else:
            runInstructions.add(instructIndex)
        # Perform the command
        if currInstruct[0] == 'acc':
            if currInstruct[1][0] == '-':
                accumulator -= int(currInstruct[1][1:])
            if currInstruct[1][0] == '+':
                accumulator += int(currInstruct[1][1:])
            instructIndex += 1
        elif currInstruct[0] == 'jmp':
            if currInstruct[1][0] == '-':
                instructIndex -= int(currInstruct[1][1:])
            if currInstruct[1][0] == '+':
                instructIndex += int(currInstruct[1][1:])
        elif currInstruct[0] == 'nop':
            instructIndex += 1

# First find index of all jmp commands
instructWIndex = list(zip(range(len(instructionsFinal)), instructionsFinal))
candidates = list(filter(lambda x: x[1][0] in ['jmp', 'nop'], instructWIndex))
# Iterate over candidates, seeing if we can correct the error and cause program to terminate correctly
for candidate in candidates:
    # Swap the instructions
    dup = deepcopy(instructionsFinal)
    if candidate[1][0] == 'nop':
        dup[candidate[0]][0] = 'jmp'
    elif candidate[1][0] == 'jmp':
        dup[candidate[0]][0] = 'nop'
    # Execute our new instruction set
    validation = execute_set_valid(dup) 
    if validation > 0:
        print(f'Accumulator at successful termination: {validation}')
        break
    
```
