```python
### Part 1
def contains_gold(key: str, level: int) -> bool:
    #print(f'{"--"*level}{key}')
    global lookupTable
    # Color has not already been searched
    if not lookupTable.get(key):
        # Now we need to reference the nestedItems - colors we will need to check
        followOnItems = list(nestedItems[key].keys())
        for item in followOnItems:
            # Item itself contains a shiny gold bag or one of it children does
            if item.strip() == 'shiny gold' or contains_gold(item.strip(), level+1):
                lookupTable[item.strip()] = True
                return True
            # Item does not contain gold, nor does any of its children
            lookupTable[item] = False
        # Update the global dictionary
        lookupTable[key] = False
        return False
    else:
        #print('Found existing values')
        return lookupTable.get(key)
```

```python
### Part 2
def contains_gold_bags(key, level):
    #print(f'{"--"*level}{key}')
    global lookupTable
    # Color has not already been searched
    if not lookupTable.get(key):
        # Now we need to reference the nestedItems - colors we will need to check
        # Initialize our starting value appropriately
        # Logic here is the "shiny gold" shouldn't count itself in the final count
        masterSum = 1 if level > 1 else 0
        for item, value in nestedItems[key].items():
            # Item itself contains a shiny gold bag or one of it children does
            if item.strip == 'shiny gold':
                print('VIOLATION!!!!!')
            tempSum = int(contains_gold_bags(item.strip(), level+1))
            masterSum += tempSum * int(value)
            lookupTable[item.strip()] = tempSum

        # Update the global dictionary
        lookupTable[key] = masterSum
        return masterSum
    else:
        #print('Found existing values')
        return lookupTable.get(key)
```
