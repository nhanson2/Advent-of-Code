```python
### Part 1
# Clean input
numbers = list(map(int, values.split('\n')))
# Device voltage
deviceVolt = max(numbers) + 3
numbers.append(deviceVolt)
# Starting outlet voltage is 0
currVoltage = 0
diff = 0
diffs = []
prevAdapter = None
for adapter in sorted(numbers):
    currDiff = adapter - currVoltage
    currVoltage = adapter
    diffs.append(currDiff)
    #if not prevAdapter:
# Sum up differences
print(f'Jolatage Distribution: {diffs.count(1) * diffs.count(3)}')
```

```python
### Part 2
from collections import defaultdict
# Let's introduce a set for ease of access
numbers = sorted(numbers)
# Insert 0 here since we are starting at the outlet
numbers.insert(0,0)
numbersSet = set(numbers)
numWays = defaultdict(lambda: 1)

def find_the_ways(number):
    # Find the number of ways to get to that number
    temp = []
    # Check if the numbers, 1, 2, or 3 below us are reachable
    if number - 1 in numbersSet:
        # Number exists! Now how many ways are there to get to it
        temp.append(numWays[number-1])
    else:
        # Number does not exist, unreachable
        numWays[number-1] = 1
    if number - 2 in numbersSet:
        temp.append(numWays[number-2])
    else:
        numWays[number-2] = 1
    if number - 3 in numbersSet:
        temp.append(numWays[number-3])
    else:
        numWays[number-3] = 1
    
    return sum(temp)


# Now go over all the numbers
for number in numbers:
    # Find number of ways to get the current number
    print(f'number: {number} -> Ways to get there: {find_the_ways(number)}')
    tWays = find_the_ways(number)
    numWays[number] = (tWays or 1)
print(f'\nNumber of possible combindations from plug to device {numWays[number]}')
```
