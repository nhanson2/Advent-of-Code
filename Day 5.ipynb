```python
# rows are numbered 0 - 127
import math
import typing
def boarding_pass_decoder(boardingPass: str) -> int:
    belief = [0, 127, 0, 7]
    for letter in boardingPass:
        midX = (belief[1] - belief[0]) / 2
        midY = (belief[3] - belief[2]) / 2
        if letter == 'F':
            # Need to update our max x belief to the midpoint
            belief[1] = math.floor(belief[1] - midX)
        elif letter == 'B':
            # Need to update our min x belief to the midpoint 
            belief[0] = math.ceil(belief[0] + midX)
        elif letter == 'L':
            belief[3] = math.floor(belief[3] - midY)
        elif letter == 'R':
            belief[2] = math.ceil(belief[2] + midY)
    return belief[0] * 8 + belief[2]


# Search through all boarding passes for the max
print(f'Highest Seat ID:{max(map(boarding_pass_decoder, values.split()))}')
```

```python
# Imports
from collections import defaultdict 

# Part 2
def boarding_pass_to_seat(boardingPass: str) -> int:
    belief = [0, 127, 0, 7]
    for letter in boardingPass:
        midX = (belief[1] - belief[0]) / 2
        midY = (belief[3] - belief[2]) / 2
        if letter == 'F':
            # Need to update our max x belief to the midpoint
            belief[1] = math.floor(belief[1] - midX)
        elif letter == 'B':
            # Need to update our min x belief to the midpoint 
            belief[0] = math.ceil(belief[0] + midX)
        elif letter == 'L':
            belief[3] = math.floor(belief[3] - midY)
        elif letter == 'R':
            belief[2] = math.ceil(belief[2] + midY)
    return (belief[0], belief[2])
# Get all seats on plane
seats = defaultdict(list)
allSeats = map(boarding_pass_to_seat, values.split())
for seatPair in allSeats:
    seats[seatPair[0]].append(seatPair[1])
# Discard first and last rows
for row in sorted(seats.keys())[1:-1]:
    if len(seats[row]) != 8:
        # We've found our target
        emptySeat =  [z for z in list(range(8)) if z not in seats[row]]
        print(f'Santa\'s seat ID: {row * 8 + emptySeat[0]}')
        break
```
