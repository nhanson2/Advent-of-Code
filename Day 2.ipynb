```python
# Clean up policies into readable format
import typing
from typing import Dict
policies = []
for policy in values:
    rule, passwd = policy.strip().split(':')
    rangeVals, targetLetter = rule.split()
    minVal, maxVal = [int(x) for x in rangeVals.split('-')]
    policies.append({
        'character': targetLetter.strip(),
        'minVal': minVal,
        'maxVal': maxVal,
        'password': passwd.strip()
    })

def count_correct(toCheck: Dict)->int:
    return int(toCheck.get('minVal') <= toCheck.get('password').count(toCheck.get('character')) <= toCheck.get('maxVal'))

print('Total Valid: {}'.format(sum(map(count_correct, policies))))
```
